```python
# Superposing mising values

for col in dataset.columns:
    if dataset[col].isna().sum() > 0:
        dataset[col] = impute_dataset[col]
```

```python
#Confirming the helath of our dataset

dataset.isna().sum()
```

```python
# COrrecting typo in feature 1 name spelling

dataset.rename(columns={'HeartDisease-Hypertension)':'HeartDisease-Hypertension'}, inplace=True)
```

```python
dataset
```

### Step 2: Data type Validation

```python
dataset.columns
```

```python
#Convertying the following features to integer
float_var = ['PhysicalHealth', 'MentalHealth', 'SleepTime' ]

for col in dataset[float_var].columns:
    dataset[col] = dataset[col].astype('int64')

#Converting BMI to float dtype
dataset['BMI'] = dataset['BMI'].astype('float64')

#COnverting other to object dtype
obj_var = ['HeartDisease-Hypertension', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'PhysicalActivity', 'Asthma', 'KidneyDisease', 'SkinCancer']

for col in dataset[obj_var].columns:
    dataset[col] = dataset[col].astype('object')

#COnverting features to categorical dtypes
cat_var = ['Sex', 'AgeCategory', 'Diabetic', 'GenHealth', 'Tribe']

for col in dataset[cat_var].columns:
    dataset[col] = dataset[col].astype('category')
```

```python
dataset.info()
```

## Feature Engineering

```python
dataset.info()
```

```python
sns.set_style('whitegrid')
sns.distplot(dataset['BMI'], kde=True, color='red', bins=40)
```

```python
bmi_status = []

for bmi in dataset['BMI'].values:
    if round(bmi,1) < 18.5:
        bmi_status.append('Underweight')
    elif (round(bmi,1) >= 18.5) &  (round(bmi,1) <= 24.9):
        bmi_status.append('Healthy Weight')
    elif (round(bmi,1) >= 25.0) &  (round(bmi,1) <= 29.9):
        bmi_status.append('Overweight') 
    elif (round(bmi,1) >= 30.0) &  (round(bmi,1) <= 34.9):
        bmi_status.append('Low Risk Obesity')
    elif (round(bmi,1) >= 35.0) &  (round(bmi,1) <= 39.9):
        bmi_status.append('MOderate Risk Obesity')
    else:
        bmi_status.append('High Risk Obesity')
```

```python
dataset['BMI Status'] = bmi_status

dataset['BMI Status'] = dataset['BMI Status'].astype('category')   #Ascribing category dtype
```
